 # 사각형 넓이 구하기

class 넓이:
  def __init__(self,밑변,높이,윗변=0,아랫변=0):
    self.밑변=밑변
    self.높이=높이
    self.윗변=윗변
    self.아랫변=아랫변
  
  def 평행사변형(self):
    return self.밑변 * self.높이

  def 사다리꼴(self):
    return (self.윗변 + self.아랫변) * self.높이/2


```python
s1= 넓이(10, 5)
print(s1.평행사변형())
s2= 넓이(10, 5, 8, 6)
print(s2.사다리꼴())
```

# 키오스크

```python
class 키오스크:
  def __init__(self):
    self.메뉴 = {"아메리카노":2000, "라떼":3000, "모카":3000, "유자차":2500, "녹차":2500, "초코라떼":3000}
    self.주문내역 = []
    self.메뉴목록 = list(self.메뉴.keys())
    self.주문금액 = []
  
  def 메뉴판(self):
    번호=1
    for 이름, 가격 in self.메뉴.items():
      print(f"{번호}.{이름}:{가격}")
      번호= 번호+1

  def 주문(self):
    while True:
      self.메뉴판()
      번호 = int(input("메뉴를 선택하세요(종료는0번)"))
      if 번호 == 0:
        break
      선택 = int(input("1.HOT 2.ICE:"))
      if 선택 == 1:
        온도="HOT"
      else:
        온도="ICE"
      self.주문금액.append(self.메뉴[self.메뉴목록[번호-1]])  
      self.주문내역.append(온도 + self.메뉴목록[번호-1])
      print(f"{온도} {self.주문내역}")
  
  def 결제(self):
    while True:
      print(f"총 금액:{sum(self.주문금액)}원")
      결재 = int(input("결제방법을 선택해주세요 1.현급 2.카드"))
      if 결재 == 1:
        print("직원에게 현금을 주고 거래를 완료해주세요")
        break
      elif 결재 == 2:
        print("카드를 넣어주세요")
        break
      else:
        print("다시 결제해주세요")

  def 주문표(self):
    print("==============================================================")
    for i in range(len(self.주문내역)):
      print(f"{self.주문내역[i]}: {self.주문금액[i]}")  
    print(f"총 금액 : {sum(self.주문금액)}원")
    print("==============================================================")
        

    

k=키오스크()
k.주문()
k.결제()
k.주문표()
```
