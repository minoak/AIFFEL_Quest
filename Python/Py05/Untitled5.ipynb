```python
#1.class생성,계좌 갯수 저장
#2.get account_num()추가
#3.입급, deposit 추가
#4.출급, withdraw 추가 계좌 잔고 이상으로 출금 불가
#6.display_info()추가 잔고는 세자리마다 쉼표
#7.입금횟수가 5번 이상일 때 1퍼센트의 이자(잔고기준)
#8.account 클래스에서 3개 이상의 인스턴스 생성, 리스트에 저장
#9.잔고가 100만원 이상인 고객의 정보 출력
#10.입출금 기록이 업데이트 되도록 해야함 deposit_history, withdraw_history추가


```

```python
import random


class Account:
  count=0
  def __init__(self, 예금주, 잔액):
    Account.count+=1
    self.은행이름= "SC은행"
    self.예금주= 예금주
    self.잔액= 잔액
    self.계좌번호 = f"{random.randint(0, 999):03d}-{random.randint(0, 99):02d}-{random.randint(0,999999):06d}"
    self.입금횟수 = 0
    self.입금내역 = []
    self.출금내역 = []


  def get_account_num(self):
    print(f"졔좌수:{Account.count}")

  def deposit(self,입금):
    if 입금 >= 1:
      self.잔액 += 입금
      self.입금횟수 += 1
      self.입금내역.append(입금)
      if self.입금횟수 == 5:
        self.잔액+= self.잔액 * 0.01

    else:
      print("입금액은 1원 이상부터 가능합니다.")

  def withdraw(self,출금):
    if 출금 <= self.잔액:
      self.잔액 -=출금
      self.출금내역.append(출금)
    else:
      print("잔고가 부족합니다.")

  def display_info(self):
    print(f"{self.은행이름},{self.예금주},{self.잔액:,},{self.계좌번호}")

  def deposit_history(self):
    print(self.입금내역)

  def withdraw_history(self):
    print(self.출금내역)


account = [
      Account("민옥", 1500000),
      Account("민수",20000000),
      Account("수지",100000000)
  ]


for i in account:
  if i.잔액 >= 1000000:
    i.display_info()

```
