```python
while True:
    try:
      a = int(input("첫 번째 숫자: "))
      b = int(input("두 번째 숫자: "))
      c = input("공식(+,-,*,/,**):")
      if c == "+":
        print(a+b)
      elif c == "-":
        print(a-b)
      elif c == "*":
        print(a*b)
      elif c == "/":
        print(a/b)
      elif c == "**":
        print(a**b)
      else:
        print("지원하지 않는 수식")
      계속 = input("계속?(y/n)")
      if 계속 == "n":
        break
    except ValueError:
      print("잘못된 값 입력")
    except ZeroDivisionError:
      print("0으로 나눌 수 없음")

```
